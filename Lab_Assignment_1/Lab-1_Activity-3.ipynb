```python
import os
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from numpy import asarray
count = 0
path = "train"
data1 = np.zeros(2353)
print(data1.shape)
for root, _, files in os.walk(path):
    current_directory_path = os.path.abspath(root)
    for f in files:
        name, ext = os.path.splitext(f)
        if ext == ".jpg":
            # Step 1: Read Image from the folder
            current_image_path = os.path.join(current_directory_path, f)
            
            # Step 2 - a: Store it in the numpy array
            current_image = Image.open(current_image_path).resize((28,28))
            data = asarray(current_image)
            myData = data.flatten()
            label = int(root[6:])
            myData = np.append(myData,[label],axis=0)
            data1 = np.vstack ((data1, myData))
data1 = np.delete(data1,0,0).astype(int)
print(data1.shape)
```

```python
from numpy import savetxt
# Step 2 - b: save to csv file
savetxt('currency.csv', data1, delimiter=',')
```

```python
import pandas as pd
df = pd.read_csv('currency.csv',header=None)
df
```

```python
X = df.drop(df.columns[2352],axis=1)
Y = df[df.columns[2352]]
```

```python
X
```

```python
print(X.shape,Y.shape)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
import seaborn as sn
gnb = GaussianNB()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.05, random_state=0)
y_pred_gnb = gnb.fit(X_train, y_train).predict(X_test)
cnf_matrix_gnb = confusion_matrix(y_test, y_pred_gnb)
print(cnf_matrix_gnb)
print(y_test.shape)
print(f1_score(y_test, y_pred_gnb, average='weighted'))
print(precision_score(y_test, y_pred_gnb, average='weighted'))
print(recall_score(y_test, y_pred_gnb, average='weighted'))
df_cm = pd.DataFrame(cnf_matrix_gnb)
plt.figure()
sn.heatmap(df_cm,vmin=0, vmax=5, annot=True)
```
