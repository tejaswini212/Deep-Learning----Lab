### Activity-1: Demonstate the following activation functions:
***1)Sigmoid\
2)Tanh\
3)ReLU\
4)Leaky - ReLU\
5)Exponential LU***

```python
# Import Necessary Libraries:
import numpy as np
from matplotlib.pyplot import plot as plt
```

Note: We will assume in this whole activity that we have an array of input to the activation function as the output of any deep learning model will be stored in a column for various input samples

```python
#1: Sigmoidal Activation Function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
y = np.arange(-20,20,1)
plt(y,sigmoid(y))
```

```python
#2: tanh Activation Function
def tanh(z):
    return np.tanh(z)

def tanh_m(z):
    return 2 / (1 + np.exp(-2*z)) -1
```

```python
y = np.arange(-20,20,1)
plt(y,tanh(y))
```

```python
plt(y,tanh_m(y))
```

```python
#3: ReLU Activation Function
def relu(z):
    return max(0, z)
```

```python
x = np.arange(-20,20,1)
y = []
for i in range(-20,20,1):
    y.append(relu(i))
plt(x,y)
```

```python
#4: Leaky - ReLU Activation Function
def l_relu(z):
    return np.maximum(0.01 * z, z)
```

```python
y = np.arange(-2,2,1)
plt(y,l_relu(y))
```

Note: We can notice a bump here where y value is less than 0 and the line is bit tilted below 0

```python
#5: Exponential Leaky - ReLU Activation Function
def elu(z,alpha):
    return z if z >= 0 else alpha*(np.exp(z) -1)
```

```python
x = np.arange(-20,20,1)
y = []
for i in range(-20,20,1):
    y.append(elu(i,3))
plt(x,y)
```
